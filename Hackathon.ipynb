```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('tracks_features.csv')
```

```python
df2 = df[['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']]
```

```python
treino = df2.iloc[0:842817,]
```

```python
teste = df2.iloc[842818:len(df2)-1,]
```

```python
inercia = []
```

```python
kmeans = KMeans(n_clusters = 10, random_state = 0, n_init='auto')
modelo = kmeans.fit(treino)
```

```python
grupos_teste = modelo.predict(teste)
```

```python
teste['grupo'] = grupos_teste
```

```python
import plotly.express as px
from sklearn.decomposition import PCA
features = teste.loc[:,['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']]
proj_2d= PCA(n_components=2).fit_transform(features)
```

```python
artist_name = "Sounds of Red Bull"
```

```python
result = sp.search(q='artist:' + artist_name, type='artist')
artist = result['artists']['items'][0]
```

```python
albums = sp.artist_albums(artist['id'], album_type='album')
```

```python
audio_features = []
```

```python
for album in albums['items']:
    album_tracks = sp.album_tracks(album['id'])
    for track in album_tracks['items']:
        
        features = sp.audio_features(track['id'])[0]
        features['track_name'] = track['name']  
        audio_features.append(features)
        if len(audio_features) >= 10:
            break  
    if len(audio_features) >= 10:
        break
```
